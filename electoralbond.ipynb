```python
import PyPDF2
import pandas as pd
```

```python
pdf_path = 'C:/Main/Main2/electoral bonds/party_data.pdf'
```

```python
def extract_text_from_all_pages(pdf_path):
    pdf_file = open(pdf_path, 'rb')
    pdf_reader = PyPDF2.PdfReader(pdf_file)
    all_page_text = ""
    for page_number in range(len(pdf_reader.pages)):
        page_text = pdf_reader.pages[page_number].extract_text()
        all_page_text += page_text + "\n"  # Add a newline between pages
    pdf_file.close()
    return all_page_text

# Extract text from the first page
page_text = extract_text_from_all_pages(pdf_path)
page_text = page_text.replace('\n', '', 1)
```

```python
# Split the text into lines
lines = page_text.split('\n')

# Split each line into columns based on spaces
data = [line.split() for line in lines[1:] if line.strip()]
```

```python
# Combine cells from 2nd to last 2nd into three cells
combined_data = [[row[0], ' '.join(row[1:-1]), row[-1]] for row in data]

# Create a DataFrame
df = pd.DataFrame(combined_data, columns=['Date of Encashment', 'Name of the Political Party', 'Denomination'])
df
```

```python
# Assuming 'df' is the DataFrame containing your data
unique_parties = set(df['Name of the Political Party'])
num_unique_parties = len(unique_parties)
print("Number of unique political parties:", num_unique_parties)
```

```python
# Replace specific party names with new names
df['Name of the Political Party'] = df['Name of the Political Party'].replace({
    'SHIVSENA (POLITICAL PARTY)': 'SHIVSENA',
    'RASTRIYA JANTA DAL': 'RASHTRIYA JANTA DAL',
    'NATIONALIST CONGRESS PARTY PARLIAMENT OF': 'NATIONALIST CONGRESS PARTY',
    'NATIONALIST CONGRESS PARTY MAHARASHTRA PRADESH': 'NATIONALIST CONGRESS PARTY',
    'PRESIDENT, ALL INDIA CONGRESS COMMITTEE': 'INDIAN NATIONAL CONGRESS',
    'YSR CONGRESS PARTY (YUVAJANA SRAMIKA R' : 'YUVAJANA SRAMIKA RYTHU CONGRESS PARTY',
    # Add more replacements as needed
})
```

```python
# Convert 'Denomination' column to numeric (remove commas first)
df['Denomination'] = df['Denomination'].str.replace(',', '').astype(int)
```

```python
# Sum the donations for each party
party_donations = df.groupby('Name of the Political Party')['Denomination'].sum()

# Sort the party donations in descending order
party_donations_sorted = party_donations.sort_values(ascending=False)

# Display the sorted total donations for each party
print(party_donations_sorted)
```

```python
# Convert 'Date of Encashment' column to datetime
df['Date of Encashment'] = pd.to_datetime(df['Date of Encashment'])

# Group the data by month, year, and party, and calculate the sum of donations
grouped_df = df.groupby([df['Date of Encashment'].dt.to_period('M'), 'Name of the Political Party'])['Denomination'].sum().unstack().fillna(0)

# Iterate over each month from April 2019 to January 2024
start_date = pd.Period('2019-04', 'M')
end_date = pd.Period('2024-01', 'M')

cumulative_donation = 0
for month_year in pd.period_range(start=start_date, end=end_date, freq='M'):
    try:
        # Get the total donation for BJP in the specified month and year
        bjp_donation = grouped_df.loc[str(month_year)]['BHARTIYA JANTA PARTY']
    except KeyError:
        # Set total donation to 0 if there is no data for that month
        bjp_donation = 0
    
    # Cumulatively add the donation amount for BJP
    cumulative_donation += bjp_donation
    
    # Display the cumulative donation for BJP after each month
    print(f"Cumulative donation for BJP after {month_year}: {cumulative_donation}")
```

```python
# Dictionary to store the cumulative donations for each party
cumulative_donations = {
    'BHARTIYA JANTA PARTY': 0, 'ALL INDIA TRINAMOOL CONGRESS': 0, 'INDIAN NATIONAL CONGRESS': 0,
    'BHARAT RASHTRA SAMITHI': 0, 'BIJU JANATA DAL': 0, 'DMK PARTY IN PARLIAMENT': 0,
    'YUVAJANA SRAMIKA RYTHU CONGRESS PARTY': 0, 'SHIVSENA': 0, 'TELUGU DESAM PARTY': 0,
    'AAM AADMI PARTY': 0, 'RASHTRIYA JANTA DAL': 0, 'NATIONALIST CONGRESS PARTY': 0,
    'JANATA DAL ( SECULAR )': 0, 'SIKKIM KRANTIKARI MORCHA': 0, 'ADYAKSHA SAMAJVADI PARTY': 0,
    'JHARKHAND MUKTI MORCHA': 0, 'JANASENA PARTY': 0, 'ALL INDIA ANNA DRAVIDA MUNNETRA KAZHAGAM': 0,
    'SHIROMANI AKALI DAL': 0, 'MAHARASHTRAWADI GOMNTAK PARTY': 0, 'GOA FORWARD PARTY': 0,
    'BIHAR PRADESH JANTA DAL(UNITED)': 0, 'SIKKIM DEMOCRATIC FRONT': 0,
    'JAMMU AND KASHMIR NATIONAL CONFERENCE': 0
}

# Dictionary to store cumulative donations after each iteration
cumulative_progression = {}

# Convert 'Date of Encashment' column to datetime
df['Date of Encashment'] = pd.to_datetime(df['Date of Encashment'])

# Group the data by month, year, and party, and calculate the sum of donations
grouped_df = df.groupby([df['Date of Encashment'].dt.to_period('M'), 'Name of the Political Party'])['Denomination'].sum().unstack().fillna(0)

# Iterate over each month from April 2019 to January 2024
start_date = pd.Period('2019-04', 'M')
end_date = pd.Period('2024-01', 'M')

for month_year in pd.period_range(start=start_date, end=end_date, freq='M'):
    month_year_str = str(month_year)
    monthly_cumulative = {}
    for party in cumulative_donations.keys():
        try:
            # Get the total donation for the party in the specified month and year
            party_donation = grouped_df.loc[month_year_str][party]
        except KeyError:
            # Set total donation to 0 if there is no data for that month
            party_donation = 0
        
        # Cumulatively add the donation amount for the party
        cumulative_donations[party] += party_donation
        # Store the cumulative donation for the party in the monthly_cumulative dictionary
        monthly_cumulative[party] = cumulative_donations[party]
    
    # Add the monthly_cumulative dictionary to the cumulative_progression dictionary
    cumulative_progression[month_year_str] = monthly_cumulative

# Display the cumulative donations for each party after each iteration
for month_year, cumulative_dict in cumulative_progression.items():
    print(f"{month_year}: {cumulative_dict}")
```

```python
import json

# Display the cumulative donations for each party after each iteration
with open('cumulative_progression.json', 'w') as file:
    json.dump(cumulative_progression, file)

print("JSON file created and downloaded.")
```
